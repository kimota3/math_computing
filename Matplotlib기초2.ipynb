## 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5)
y = 3*x +1

plt.plot(x,y)
plt.show()
```

## 꾸미기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5)# 정의역 설정


y1 = 3*x   # 함수 설정
y2 = 2*x

plt.plot(x,y1)
plt.plot(x,y2,linestyle="dashed")
```

```python
plt.grid() # grid(눈금,격자)표시
```

```python
plt.plot(x,y1)
plt.plot(x,y2, label="y2", linestyle="dashed")#범례와 성의 스타일
plt.legend()#범례 표시

plt.show
```

```python
plt.title("Graph")


plt.xlabel("X", size=12)
plt.ylabel("Y", size=12)

plt.grid()

plt.plot(x,y1,label="y1" )
plt.plot(x,y2, label="y2", linestyle="dashed")
plt.legend()

plt.show
```
