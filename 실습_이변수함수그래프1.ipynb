### .plot_surface 로 그리기

```python
# 2변수 함수 그래프

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


ax3 = plt.axes(projection='3d') #3차원 축


x = np.linspace(-3, 3, 1000) # 정의역 정의(x)
y = np.linspace(-3, 3, 1000) # 정의역 정의(y)


X, Y = np.meshgrid(x, y) # 정의역의 meshgrid

fXY=(2*X**2+Y**2)*np.e**(-X**2-Y**2) #함수(치역)

ax3.plot_surface(X, Y, fXY) #그래프 만들기


plt.show()
```

## 그래프 꾸미기

- 그래프 제목
- 그래프 크기
- 보이는 각도 조절

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(8,8)) # 크기

ax3 = plt.axes(projection='3d') 


x = np.linspace(-3, 3, 101) 
y = np.linspace(-3, 3, 101) 


X, Y = np.meshgrid(x, y) 

fXY=(2*X**2+Y**2)*np.e**(-X**2-Y**2) 

ax3.plot_surface(X,Y,fXY,rstride = 2,cstride = 20,alpha = 0.4,color = 'gray',edgecolor = 'blue')
 #(X,Y,fXY,열선,행선(1:전부표시, 2:한칸씩 띄워서표시), 투명도, 색깔, 색깔)

plt.title("$z=(2x^2+y^2)e^{-x^2-y^2}$",fontsize=15) #제목  

ax3.xaxis.set_tick_params(labelsize=5) #x축 눈금
ax3.yaxis.set_tick_params(labelsize=10)
ax3.zaxis.set_tick_params(labelsize=15)
ax3.set_xlabel(r'$x$', fontsize =10) #x축
ax3.set_ylabel(r'$y$', fontsize =10)
ax3.set_zlabel(r'$z$', fontsize =20)
#ax3.set_xlabel('x', fontsize =20) # x축
#ax3.set_ylabel('y', fontsize =20)  # y축

ax3.view_init(15,60) # 상하회전각도 ( 0 - 옆에서 본 각도, 90 - 위에서 본 각도)
                     # 좌우회전각도 ( 양수 - 시계방향 회전, 음수 - 시계반대방향 회전)

plt.show()
```

# 등위선 그리기

## 이변수 함수의 등위선 
- contour() 함수 이용
- level 설정 및 주석 달아주기

```python
import numpy as np
import matplotlib.pyplot as plt

def j(x,y):
    z=x**2 - y**2
    return z

x_2=np.linspace(-1,1,21)
y_2=np.linspace(-1,1,21)

X_2,Y_2 = np.meshgrid(x_2,y_2)

Z_2 = j(X_2,Y_2)#함수

CC = plt.contour(X_2,Y_2,Z_2,levels=np.arange(-1,1,0.2),cmap=plt.cm.rainbow)#등위선(---,level설정,--)
plt.clabel(CC,inline=2,fontsize = 9) #레벨주석 달아주기
plt.colorbar()# 컬러바 생성 
plt.show()
```

## 이변수 함수의 등위선과 그래디언트
- sympy 사용
- diff(), subs()
- ravel()함수는 다차원 배열을 1차원 배열로 바꾸어 준다.  
- zip()함수는 동일한 개수로 이루어진 자료형을 묶어 주는 역할을 함.
- quiver() 함수는 벡터장을 그려주는 함수.   

```python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
#f = x**2 - y**2 
k =10*x**2 - 10*y**2 

k_by_x = sp.diff(k, x)
k_by_y = sp.diff(k, y)

n = 21
x_3 = np.linspace(-0.5, 0.5,n)
y_3 = np.linspace(-0.5, 0.5, n)

X_3, Y_3= np.meshgrid(x_3, y_3)

Z_3 = [float(k.subs(x,xv).subs(y,yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())] # 함수
Z_3=np.array(Z_3).reshape(n,n)
#print(Z_3)

fig=plt.figure(figsize=(16, 8))

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_3, Y_3, Z_3, cmap='viridis')
ax.view_init(10,70) 
plt.show()

plt.contour(X_3, Y_3, Z_3, 50, levels = np.arange(-1, 1, 0.2), cmap=plt.cm.rainbow)#등위선 

X_4, Y_4 = np.meshgrid( np.linspace(-0.5,0.5, 11), np.linspace(-0.5, 0.5, 11) )

XG = [float(k_by_x.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

YG = [float(k_by_y.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

plt.quiver(X_4, Y_4, XG, YG, width=0.005, scale=100, color='gray') # 그래디언트 생성
plt.show()
```

```python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
k = x**2 + y**2 + x*y - sp.cos(x)*5

k_by_x = sp.diff(k, x)
k_by_y = sp.diff(k, y)

n=50
x_3 = np.linspace(-10, 10, n)
y_3 = np.linspace(-10, 10, n)
X_3, Y_3= np.meshgrid(x_3, y_3)

Z_3 = [float(k.subs(x,xv).subs(y,yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
Z_3=np.array(Z_3).reshape(n,n)

fig=plt.figure(figsize=(7, 5))

plt.contour(X_3, Y_3, Z_3, 50, levels = np.logspace(-1.0, 2.0, 20), cmap=plt.cm.rainbow)

X_4, Y_4 = np.meshgrid( np.linspace(-10, 10, 10), np.linspace(-10, 10, 10) )

XG = [float(k_by_x.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

YG = [float(k_by_y.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

plt.quiver(X_4, Y_4, XG, YG, width=0.005, scale=500, color='red')
plt.show()
```

# 라그랑즈 승수법

- 주어진 조건을 만족하는 목적함수의 최댓값과 최솟값을 알아본다.
- 여기서 조건은 방정식으로 주어진다. 
- 이러한 상황을 그림으로 구현한다. 

[예제]

(1) $x_1 + x_2 = 1$의 조건에서 $x_1^2 + x_2^2$ 의 최댓값과 최솟값을 알아보자. 

(2) $x_1^2 + x_2^2 = 1$ 의 조건에서 $x_1^2 + x_2$ 의 최댓값과 최솟값을 알아보자. 

```python
import numpy as np
import matplotlib.pyplot as plt


def f1(x1, x2):                     # 목적함수 f(x) = x1^2 + x2^2
    return x1**2 + x2** 2

x1 = np.linspace(-3.5, 3.5, 100)
x2 = np.linspace(-3.5, 3.5, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f1(X1, X2)

x2_g = 1 - x1                       # 등식 제한조건 방정식 g(x) = x1 + x2 - 1 = 0

fig=plt.figure(figsize=(6, 6))
CC=plt.contour(X1, X2, Y, colors="gray",levels=[0.2,0.5, 2,4,6])
plt.plot(x1, x2_g, 'g-')

plt.clabel(CC,inline=2,fontsize = 9)

plt.plot([0.5], [0.5], 'bo', ms=10)

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def f1(x1, x2):                 # 목적함수 f(x) = x1^2 + x2
    return x1 ** 2 + x2

x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f1(X1, X2)

x2_g = np.sqrt(+1 - x1**2)     # 등식 제한조건 방정식   x1**2 + x2**2 = 1
x3_g = -np.sqrt(+1 - x1**2)

fig=plt.figure(figsize=(6, 6))
CC=plt.contour(X1, X2, Y, colors="gray", levels=[-1,0.5, 5/4,2, 8])
plt.clabel(CC,inline=2,fontsize = 9)

plt.colorbar
plt.plot(x1, x2_g, 'g-')
plt.plot(x1, x3_g, 'g-')

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.show()
```
