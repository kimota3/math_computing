```python
import numpy as np
import matplotlib.pyplot as plt

# 목적함수 f(x) = x1^2 + x2^2
def f1(x1, x2):
    return x1 ** 2 + x2
x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f1(X1, X2)

# 등식 제한조건 방정식   x1**2 + x2**2 = 1
x2_g = np.sqrt(+1 - x1**2)
x3_g = -np.sqrt(+1 - x1**2)



fig=plt.figure(figsize=(6, 6))
CC=plt.contour(X1, X2, Y, colors="gray", levels=[-1,0.5, 5/4,2, 8])
plt.clabel(CC,inline=2,fontsize = 9)

plt.colorbar
plt.plot(x1, x2_g, 'g-')
plt.plot(x1, x3_g, 'g-')

#plt.plot([], [], 'rP')
#plt.plot([], [], 'ro', ms=10)

#plt.xlim(-3, 1)
#plt.ylim(-1, 1)
#plt.xticks(np.linspace(-3, 3, 9))
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
#plt.title(" ")
plt.show()
```
