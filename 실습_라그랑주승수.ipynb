```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x$', size =12)
ax.set_ylabel(r'$y$', size =12)
ax.set_zlabel(r'$z$', size =12)

def f(x,y):
    z= -(x+1)*(x)*(x-1)*(x-2)-(y+1)*(y)*(y-1)*(y-2)
    #z= -(x+1)*(x)*(x-1)*(x-2)-(y+1)*(y)
    return z

## (x1, x2)
x= np.linspace(-1.2, 2) 
y= np.linspace(-1, 2)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1,X2)
#Z = 50 * (X2 - X1 **2) **2 + (2 - X1)**2

ax.scatter3D(X1, X2, Z, marker='*', color='blue')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def j(x,y):
    z=x**2 - y**2
    return z

x_2=np.linspace(-1,1,21)
y_2=np.linspace(-1,1,21)

X_2,Y_2 = np.meshgrid(x_2,y_2)

#print(X_2)
#print(Y_2)

Z_2 = j(X_2,Y_2)

CC = plt.contour(X_2,Y_2,Z_2,levels=np.arange(-1,1,0.2),cmap=plt.cm.rainbow)

plt.clabel(CC,inline=2,fontsize = 9)

plt.colorbar

plt.show()
```

```python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
#f = x**2 + y**2 +x*y
k =10*x**2 - 10*y**2 
#j = x**2 - y**2

#def k(x,y):
 #   z = x**2 + y**2 + x*y - sp.sin(x)*5
  #  return z

k_by_x = sp.diff(k, x)
k_by_y = sp.diff(k, y)

# make meshgrid with xs, ys 
n = 21
x_3 = np.linspace(-0.5, 0.5,n)
y_3 = np.linspace(-0.5, 0.5, n)
X_3, Y_3= np.meshgrid(x_3, y_3)

# make zs 
#zs = [float(k.subs(x, xv).subs(y, yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
#zs = np.array(zs).reshape(sample_size, sample_size)
Z_3 = [float(k.subs(x,xv).subs(y,yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
Z_3=np.array(Z_3).reshape(n,n)
#print(Z_3)

fig=plt.figure(figsize=(16, 8))

#fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_3, Y_3, Z_3, cmap='viridis')
ax.view_init(10,70) 
plt.show()

#ax=fig.add_subplot(121)
plt.contour(X_3, Y_3, Z_3, 50, levels = np.arange(-1, 1, 0.2), cmap=plt.cm.rainbow)

# gradient

X_4, Y_4 = np.meshgrid( np.linspace(-0.5,0.5, 11), np.linspace(-0.5, 0.5, 11) )

XG = [float(k_by_x.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

YG = [float(k_by_y.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

plt.quiver(X_4, Y_4, XG, YG, width=0.005, scale=100, color='gray')
plt.show()
```

```python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
k = x**2 + y**2 + x*y - sp.cos(x)*5

k_by_x = sp.diff(k, x)
k_by_y = sp.diff(k, y)

# make meshgrid with xs, ys 
n=50
x_3 = np.linspace(-10, 10, n)
y_3 = np.linspace(-10, 10, n)
X_3, Y_3= np.meshgrid(x_3, y_3)

# make zs 
#zs = [float(k.subs(x, xv).subs(y, yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
#zs = np.array(zs).reshape(sample_size, sample_size)
Z_3 = [float(k.subs(x,xv).subs(y,yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
Z_3=np.array(Z_3).reshape(n,n)
#print(Z_3)

fig=plt.figure(figsize=(8, 6))

#fig=plt.figure()
#ax = fig.add_subplot(121, projection='3d')
#ax.plot_surface(X_3, Y_3, Z_3, cmap='viridis')
#ax.view_init(10,70) 
#plt.show()

#ax=fig.add_subplot(121)
plt.contour(X_3, Y_3, Z_3, 50, levels = np.logspace(-1.0, 2.0, 20), cmap=plt.cm.rainbow)

# gradient

X_4, Y_4 = np.meshgrid( np.linspace(-10, 10, 10), np.linspace(-10, 10, 10) )

XG = [float(k_by_x.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

YG = [float(k_by_y.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

plt.quiver(X_4, Y_4, XG, YG, width=0.005, scale=500, color='red')
plt.show()
```
